```python
ipython_analysis_df =  pd.DataFrame(columns=['Description','Package/Module', 'Usage'])

for py_file in tqdm(glob.glob('ipython_files/*.py')):
    
    with open(py_file, 'r',encoding='utf-8') as file:
        source = file.read()

    for pos in range(1, len(source.splitlines())+1):
        
        try:

            script = jedi.Script(source,pos)
            d = script.goto_definitions()[0]

            temp_df = pd.DataFrame([[d.description,d.full_name, source.splitlines()[pos-1]]], columns=['Description','Package/Module', 'Usage'])
            ipython_analysis_df = ipython_analysis_df.append(temp_df, ignore_index=True)
            
        except:
            continue
```

## Extracting repos with files containing scikit learn

```python
query = """\
SELECT
    DISTINCT(repo_name)
FROM
    `Odyssey_github_sklearn.content_py_full` 
WHERE
    (REGEXP_CONTAINS(path,"sklearn")) OR
    (REGEXP_CONTAINS(repo_name,"scikit-learn")) OR
    (REGEXP_CONTAINS(repo_name,"sklearn")) OR
    (REGEXP_CONTAINS(path,"scikit-learn"))
"""
gp = GithubPython()



excludedRepos = [repo_name[0] for repo_name in gp.run(query, project="odyssey-193217")]
print(excludedRepos)
```

```python
def excludeByRepoFull(excludeByRepos):
    if not excludedRepos:
        return ""
    s = '(NOT REGEXP_CONTAINS(repo_name,"%s"))' % excludedRepos[0]
    for repo in excludedRepos[1:]:
        s += ' AND '
        s += '(NOT REGEXP_CONTAINS(repo_name,"%s"))' % repo
    return s
```

## Extracting repos with other module usages

```python
SklearnMultipleLibraries= []

for moduleName in otherModuleList:

    sklearnWithOtherModulesQuery = """\
    SELECT repo_name, COUNT(*) count
    FROM
          `Odyssey_github_sklearn.content_py_full`
          

          WHERE ((REGEXP_CONTAINS(content, 'sklearn')) AND
          (REGEXP_CONTAINS(content, '%s'))) AND  (%s)
          
          
    GROUP BY 1
    ORDER BY count DESC
    """ %(moduleName, excludeByRepoFull(excludedRepos))
    
    
    sklearnWithOtherModules = gp.run(sklearnWithOtherModulesQuery)
    
    SklearnMultipleLibraries.append(sklearnWithOtherModules)
```

```python
SklearnMultipleLibraries = pd.read_csv("sklearn_Multiple_Libaries.csv")
module_quanity = []
for i in range(17):
    print(10574 - SklearnMultipleLibraries.iloc[i].isnull().sum() - 1)
    module_quanity.append(10574 - SklearnMultipleLibraries.iloc[i].isnull().sum() - 1)
```

## Analysis of other module usages with scikit- learn

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

fig, ax = plt.subplots(figsize=(15,10))

y_pos = np.arange(len(otherModuleList))

error = np.random.rand(len(otherModuleList))
ax.barh(y_pos, module_quanity, xerr=error, align='center',
        color='green', ecolor='black')
ax.set_yticks(y_pos)
ax.set_yticklabels(otherModuleList)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_ylabel('Other Python Modules')
ax.set_xlabel('Module usage with sklearn')

plt.show()
```

```python
allWithSklearnQuery = """\
    SELECT repo_name, line
    FROM (
          SELECT DISTINCT(line), repo_name, content
          FROM(
          SELECT (SPLIT(content, '\\n')) as lines, repo_name, content
          FROM
          `Odyssey_github_sklearn.content_py_full`
          ), UNNEST(lines) line 

          WHERE (REGEXP_CONTAINS(line, 'import+ ')) AND %s
         )

        WHERE (REGEXP_CONTAINS(content, 'from sklearn+  import') OR REGEXP_CONTAINS(content, 'import sklearn+ '))  
          
         GROUP BY 1,2
    
    

    """ %(excludeByRepoFull(excludedRepos))

allWithSklearn =gp.run(allWithSklearnQuery)
```
