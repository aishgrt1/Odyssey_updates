## Cleaning ipython notebooks to be parsed

```python
ipynb_pipeline_df = pd.DataFrame(columns=['Description','Package/Module', 'Usage'])

for index in tqdm(ipynb_pipeline_files):
    
    py_file = "ipython_files/" +"file"+str(index)+".py"
    with open(py_file, 'r',encoding='utf-8') as file:
        file_content = file.read()
        
        
    file_content = re.sub(r"\,\s*\r\n", ", ", file_content)
    file_content = re.sub(r"\r\n\s+", "\r\n", file_content)
    file_content = re.sub(r"\r\n\)", ")", file_content)
    file_content = re.sub(r"\r\n}", "}", file_content)
    file_content = re.sub(r"\r\n]", "]", file_content)
    file_content = re.sub(r"\(\r\n", "(", file_content)
    file_content = re.sub(r"\[\r\n", "[", file_content)
    file_content = re.sub(r"{\r\n", "{", file_content)
    file_content = re.sub(r"\[\s*\n", "[", file_content)
    file_content = re.sub(r"\n.\s*.]", "]", file_content)
    file_content = re.sub(r"\(.\s*.\n", "(", file_content)
    file_content = re.sub(r"\n.\s*.\)", ")", file_content)
    file_content = re.sub(r"{.\s*.\n", "{", file_content)
    file_content = re.sub(r"\n.\s*.}", "}", file_content)

    
        
   
    source = file_content
    for pos in range(1, len(source.splitlines())+1):

        try:

            script = jedi.Script(source,pos)
            d = script.goto_definitions()[0]

            temp_df = pd.DataFrame([[d.description,d.full_name, source.splitlines()[pos-1]], index], columns=['Description','Package/Module', 'Usage'])
            ipynb_pipeline_df = ipynb_pipeline_df.append(temp_df, ignore_index=True)
        except:
            continue
```
